```python
def tryModel(Data, prediction ):
    #prediction = model.predict(X_test)
    #prediction.shape
    width_in_inches = 20
    height_in_inches = 10
    dots_per_inch = 70

    plt.figure(
        figsize=(width_in_inches, height_in_inches),
        dpi=dots_per_inch)

    MSE = []
    for i in range(47):
        MSE.append(mean_absolute_error(Y_test[i,:], prediction[i,:]))
    plt.plot(MSE, label=" Mean square Error", color='red')
    plt.plot(np.mean(Data, axis=0), label="Mean", color='blue')
    plt.xticks(np.arange(47), DatasetTB1.columns, rotation='vertical', fontsize=20)
    plt.text(10, 1.75, 'The Overall mean square error\n is ' + str(mean_squared_error(Y_test,prediction)) +' .' ,
         horizontalalignment='center',
         verticalalignment='top',
         multialignment='center',fontsize=20, color='green' )
    plt.grid(True)
    # Tweak spacing to prevent clipping of tick-labels
    plt.subplots_adjust(bottom=0.15)
    plt.xlabel('counties ', fontsize=20)
    plt.ylabel('values ', fontsize=20)
    plt.title('Plot of mean square error per county compared to the mean per county.', fontsize=20)
    # show a legend on the plot
    plt.legend(fontsize=20)
    # Display a figure.
    plt.savefig('Output.png')
    plt.show()
    print("******")
```

```python
prediction = model.predict(X_test)
tryModel(Data,prediction)
```

```python
names = ["RandomForestRegressor", "KNeighborsRegressor", "LinearRegression", "RidgeCV",
         "Lasso"]

regressors = [
    RandomForestRegressor(max_depth=5, random_state=0,n_estimators=100),
    KNeighborsRegressor(n_neighbors=2),
    LinearRegression(),
    RidgeCV(alphas=[1e-4,1e-3, 1e-2, 1e-1, 1]),
    Lasso(alpha=0.1)
    ]

i = 1
for name, clf in zip(names, regressors):
    n= name+".pkl"
    print('************************')
    print('*****model : ' + name + '********' )
    clf.fit(X_train, Y_train)
    joblib.dump(clf,n)
    pred = clf.predict(X_test) 
    tryModel(Data,pred )
    print(" Mean MSE : ", mean_squared_error(Y_test,pred))
```
